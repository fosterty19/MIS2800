```python
import sqlite3

# Connect to (or create) the database
conn = sqlite3.connect('NFL_Scouting.db')

# Create a cursor object to run SQL commands
cursor = conn.cursor()

print("Database connected successfully")
```

```python
# Create scouting_stats table
cursor.execute("""
DROP TABLE IF EXISTS scouting_stats;
""")
cursor.execute("""
CREATE TABLE IF NOT EXISTS scouting_stats (
  team_name TEXT,
  city TEXT,
  wins INTEGER,
  points_per_game REAL
)
""")

conn.commit()
print("Table created successfully")
```

```python
# Insert sample NFL team data
cursor.execute("""
INSERT INTO scouting_stats (team_name, city, wins, points_per_game)
VALUES
  ('Chiefs', 'Kansas City', 11, 29.2),
  ('Bills', 'Buffalo', 10, 26.5),
  ('Cowboys', 'Dallas', 12, 27.8),
  ('Jets', 'New York', 7, 18.6)
""")

conn.commit()
print("Data inserted successfully")
```

```python
cursor.execute("""
SELECT * FROM scouting_stats
WHERE points_per_game > 24
""")

results = cursor.fetchall()

print("Teams averaging more than 24 points per game:")
for row in results:
  print(row)
```

```python
cursor.execute("""
SELECT * FROM scouting_stats
ORDER BY wins DESC
""")

results = cursor.fetchall()
print("Teams sorted by wins (highest to lowest):")
for row in results:
  print(row)
```

```python
cursor.execute("""
SELECT * FROM scouting_stats
WHERE wins > 5
ORDER BY city ASC
""")

results = cursor.fetchall()

print("Teams with more than 5 wins, sorted by city:")
for row in results:
  print(row)
```

```python
conn.close()
print("Database connection closed")
```
